```
# Importing libraries
import pandas as pd
import PyPDF2
import textract
import re
```

```
# Downloading the pdf
!curl -O http://ceo.karnataka.gov.in/finalrolls_2020/English/MR/AC163/S10A163P1.pdf
```

```
# Opening file and creating instance of PdfFileReader
file = "S10A163P1.pdf"
obj = open(file, 'rb')
reader = PyPDF2.PdfFileReader(obj)
```

```
# Extracting text using PyPDF2
text = ""
for page in range(2, reader.numPages):
    text += reader.getPage(page).extractText()

# If no text is extracted using reader object, then textract is used
if text == "":
    text = textract.process(file, method='tesseract', language='eng')
```

```
# Decoding the text string to 'utf-8' encoding
text = text.decode('utf-8')
print(text)
```

```
# Spliting the large text string into list of smaller strings using "splitter_str" present in "text" string
splitter_str = "\n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n \n\n"
txt_list = text.split(splitter_str)
txt_list
```

```
# Removing the unnecessary strings from txt_list
pages = []                                  # final list containing clean data
for i in range(1, len(txt_list)):           # txt_list[0] is not needed, so starting range with 1
    if len(txt_list[i]) <= 100:             # strings with length less than 100 do not contain relevant data
        continue
    txt_list[i] = txt_list[i].lower()
    txt_list[i] = re.sub('\n', ' ', txt_list[i])            # replacing '\n' with ' '
    txt_list[i] = re.sub("name[ ]*[:;.,-=]*", "name: ", txt_list[i])    # string containing "name" followed by 0 or more spaces followed by [:;.,-=] are changed to "name:"
    txt_list[i] = re.sub("'s name", "'s_name", txt_list[i])         # joining strings such as "husbands's", "father's", "mother's" with string "name" by underscore
    txt_list[i] = re.sub("house number", "house_number", txt_list[i])   # joining "house" and "number" by underscore
    txt_list[i] = re.sub("modified in supplement date of publication.*", "", txt_list[i])   # removing unnecessary end part of string as it may create ambiguity
    txt_list[i] = txt_list[i].strip()
    pages.append(txt_list[i])

print(f"Length of list 'pages': {len(pages)}")
pages
```

```
names = []          # will store all extracted names
for page in pages:
    # res stores the list of tuples, with each tuple containing three required names and an unwanted string (either " husband's_name" or " father's_name")
    res = re.findall("%s(.*?)%s(.*?)%s(.*?)( husband's_name:| father's_name)" % (" name:", " name:", " name:"), page)
    for each_tuple in res:
        names += each_tuple[0:3]
    names = [name.strip() for name in names]

print(names)
```

```
df = pd.DataFrame()         # will contain firstname, middlename, lastname of every person

for name in names:
    name_parts = re.split("[ -\.\']\s*", name)
    length = len(name_parts)
    if length == 1:                         # Name containing only firstname
        df = df.append([[name_parts[0], None, None]])
    elif length == 2:                       # Name containing firstname and lastname
        df = df.append([[name_parts[0], None, name_parts[1]]])
    else:                                   # Name containing all three parts of name
        df = df.append( [[ " ".join(name_parts[0:length-2]), name_parts[length-2], name_parts[length-1] ]])

df.rename(columns={0:"First Name", 1:"Middle Name", 2:"Last Name"}, inplace=True)
df.reset_index(drop=True, inplace=True)

print(df)
```

```
df.to_csv("result.csv")
```
